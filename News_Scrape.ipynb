```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException


# Specify the keywords you want to search for
search_keywords = ["Oil", "Nigeria"]

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.get('https://www.reuters.com/news/archive/businessnews?view=page&page=5&pageSize=10')

# Function to handle cookies pop-up if it appears
def handle_cookies_popup():
    try:
        cookies_popup = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.ID, "onetrust-banner-sdk"))
        )
        accept_button = cookies_popup.find_element(By.ID, "onetrust-accept-btn-handler")
        accept_button.click()
    except:
        pass

# Define the number of pages you want to scrape
num_pages_to_scrape = 5
count = 0
articles = []  # Store both headline and date as tuples




# Function to check if any of the keywords are present in the text
def contains_keywords(text):
    if not search_keywords:  # If search_keywords is empty, return False
        return False
    for keyword in search_keywords:
        if keyword in text:
            return True
    return False


def find_keywords(text):
    found_keywords = []
    for keyword in search_keywords:
        if keyword in text:
            found_keywords.append(keyword)
    return found_keywords

while count < num_pages_to_scrape:
    # Handle the cookies pop-up
    handle_cookies_popup()

    # Find the "Load More" button
    loadMoreButton = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CLASS_NAME, "control-nav-next"))
    )

    

    # Get the number of articles before clicking the "Load More" button
    num_articles_before_load = len(articles)

    

    # Get the headlines and dates for the new articles
    news_headlines = driver.find_elements(By.CLASS_NAME, "story-title")
    news_dates = driver.find_elements(By.CLASS_NAME, "timestamp")

    # Calculate the new minimum length after loading more articles
    # min_length = min(len(news_headlines), len(news_dates))

    # Loop through the headlines and dates and add them as tuples to the articles list
    for i in range(min(len(news_headlines), len(news_dates))):
        headline = news_headlines[i].text
        date = news_dates[i].text

        # Find keywords in the headline and date
        headline_keywords = find_keywords(headline)
        date_keywords = find_keywords(date)

        if headline_keywords or date_keywords:
            
            articles.append((headline, date))
            print("Headline:", headline)
            print("Date:", date)

            

            if search_keywords == [""]:
                pass
            else:
                if headline_keywords:
                    print("Keywords found in headline:", ", ".join(headline_keywords))

                if date_keywords:
                    print("Keywords found in date:", ", ".join(date_keywords))

                print("Number of keywords found:", len(headline_keywords) + len(date_keywords))
                print()  # Add an empty line for better readability

    count += 1
    # Scroll to the "Load More" button
    driver.execute_script("arguments[0].scrollIntoView(true);", loadMoreButton)
    # Click the "Load More" button using execute_script
    driver.execute_script("arguments[0].click();", loadMoreButton)
    
    



# Print the total number of articles
print("Total articles:", len(articles))
print("Total number of pages", count)


# Close the browser after scraping
driver.quit()
```
